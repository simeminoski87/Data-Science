```python
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
```

```python
requests.packages.urllib3.disable_warnings()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
snapshot_url='https://www.cnbc.com/finance/'
```

```python
snapshot=requests.get(snapshot_url)
snapshot
```

```python
type(snapshot)
```

```python
raw_html=snapshot.text
print(raw_html)
```

```python
type(raw_html)
```

```python
soup=BeautifulSoup(raw_html,'html.parser')
```

```python
soup.select_one('title')
```

```python
card_titles=soup.select('.Card-title')
```

```python
card_titles
```

```python
print(card_titles)
```

```python
HTML(card_titles[0].prettify())
```

```python
type(card_titles[0])
```

```python
card_titles[4].text
```

```python
titles=[]
for card in card_titles:
  titles.append(card.text)
print(titles)
```

```python
card_times=soup.select('.Card-time')
card_times
```

```python
card_times[0].text
```

```python
from datetime import date
times=[]
for time in card_times:
  if 'hour' in time:
    times.append(date.today().strftime("%b %d %y"))
  else:
    times.append(time.text)
times
```

```python
len(titles) == len(times)
```

```python
card_classes=soup.select('.Card-eyebrow')
card_classes
```

```python
len(titles) == len(card_classes)
```

```python
classes=[]
for card in card_classes:
  classes.append(card.select_one('div').text)
classes
```

```python
classes=['Top News','Top News','Top News','Top News','Top News'] + classes
```

```python
len(classes)==len(titles)
```

```python
matrix=[]
for i in range(0,len(titles)):
  matrix.append([titles[i],classes[i],times[i]])
matrix
```

```python
df2=pd.DataFrame(np.array(matrix),columns=['title','class','time'])
```

```python
df2
```

```python
matrix2= []
for i in range(0,len(titles)):
  node={}
  node['title']=titles[i]
  node['class']=classes[i]
  node['time']=times[i]
  matrix2.append(node)
matrix2
```

```python
df22=pd.DataFrame(matrix2)
```

```python
df22
```
